### Object Class Codes

| Code | Name                     |   Kind   | Description                                                                                                                  |
|:----:|--------------------------|:--------:|------------------------------------------------------------------------------------------------------------------------------|
|  IEO | Atira                    | asteroid | An asteroid orbit contained entirely within the orbit of the Earth (Q &lt; 0.983 au). Also known as an Interior Earth Object |
|  ATE | Aten                     | asteroid | Near-Earth asteroid orbits similar to that of 2062 Aten (a &lt; 1.0 au; Q &gt; 0.983 au)                                     |
|  APO | Apollo                   | asteroid | Near-Earth asteroid orbits which cross the Earth‚Äôs orbit similar to that of 1862 Apollo (a &gt; 1.0 au; q &lt; 1.017 au)     |
|  AMO | Amor                     | asteroid | Near-Earth asteroid orbits similar to that of 1221 Amor (1.017 au &lt; q &lt; 1.3 au)                                        |
|  MCA | Mars-crossing Asteroid   | asteroid | Asteroids that cross the orbit of Mars constrained by (1.3 au &lt; q &lt; 1.666 au; a &lt; 3.2 au)                           |
|  IMB | Inner Main-belt Asteroid | asteroid | Asteroids with orbital elements constrained by (a &lt; 2.0 au; q &gt; 1.666 au)                                              |
|  MBA | Main-belt Asteroid       | asteroid | Asteroids with orbital elements constrained by (2.0 au &lt; a &lt; 3.2 au; q &gt; 1.666 au)                                  |
|  OMB | Outer Main-belt Asteroid | asteroid | Asteroids with orbital elements constrained by (3.2 au &lt; a &lt; 4.6 au)                                                   |
|  TJN | Jupiter Trojan           | asteroid | Asteroids trapped in Jupiter‚Äôs L4/L5 Lagrange points (4.6 au &lt; a &lt; 5.5 au; e &lt; 0.3)                                 |
|  AST | Asteroid                 | asteroid | Asteroid orbit not matching any defined orbit class                                                                          |
|  CEN | Centaur                  |          | Objects with orbits between Jupiter and Neptune (5.5 au &lt; a &lt; 30.1 au)                                                 |
|  TNO | TransNeptunian Object    |          | Objects with orbits outside Neptune (a &gt; 30.1 au)                                                                         |
|  PAA | Parabolic ‚ÄúAsteroid‚Äù     |          | ‚ÄúAsteroids‚Äù (objects other than comets) on parabolic orbits (e = 1.0)                                                        |
|  HYA | Hyperbolic ‚ÄúAsteroid‚Äù    |          | ‚ÄúAsteroids‚Äù (objects other than comets) on hyperbolic orbits (e &gt; 1.0)                                                    |

Orbit class descriptions above use the following symbols to indicate orbit parameters:

- e - eccentricity
- q - perihelion distance
- a - semimajor axis
- Q - aphelion distance
- P - orbital period
- Tj - Jupiter Tisserand parameter
- aJ - Jupiter nominal semimajor axis
